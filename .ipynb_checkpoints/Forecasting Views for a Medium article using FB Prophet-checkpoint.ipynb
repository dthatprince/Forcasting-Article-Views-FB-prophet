## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("Solarize_Light2")# for pretty graphs

from prophet import Prophet
```

## Importing the dataset

```python
data = pd.read_excel('medium_stats.xlsx')
data.head()
```

## Analysing the datatypes

```python
data.dtypes
```

## Plotting to get insights

```python
plt.figure(figsize=(15,8))
plt.plot(data.set_index('Date'))
plt.legend(['Views'])
```

## Converting Views column to its log value

```python
data['Views'] = np.log(data['Views'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(data.set_index('Date'))
plt.legend(['Views'])
```

## Setting Column Names

Making the dataset 'Prophet' compliant.

```python
data.columns = ['ds','y']
data.head()
```

### Making Predictions

Prophet follows the sklearn model API wherein an instance of the Prophet class is created and then the fit and predict methods are called. 

The model is instantiated by a new Prophet object and followed by calling its fit method and passing in the historical dataframe.

```python
prophet = Prophet(daily_seasonality=True)
prophet.fit(data)
```

Prophet will by default fit weekly and yearly seasonalities if the time series is more than two cycles long. It will also fit daily seasonality for a sub-daily time series. You can add other seasonalities (monthly, quarterly, hourly) if required.

```python
future = prophet.make_future_dataframe(periods=60)
forecast = prophet.predict(future)
```

```python
forecast.tail().T
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

Reverting to the original scale: Converting the logged Views back to their original values.

```python
np.exp(forecast[['yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
type(forecast)
```

## Plotting the Forecast

```python
prophet.plot(forecast);
```

```python
prophet.plot_components(forecast, figsize=(10,10));
```

## Holiday Effect

```python
articles = pd.DataFrame({
  'holiday': 'Article_Published',
  'ds': pd.to_datetime(['2018-07-02', '2018-07-06', '2018-07-08',
                        '2018-07-09', '2018-07-12', '2018-07-19', '2018-07-26', '2018-07-31',
                        '2018-08-06', '2018-08-15', '2018-07-19', '2018-08-26', '2018-08-31',
                        '2018-09-01', '2018-09-04', '2018-09-11', '2018-09-17', '2018-09-23',
                        '2018-10-02', '2018-10-09', '2018-10-18', '2018-10-19', '2018-10-26',
                        '2018-11-02', '2018-11-08', '2018-11-24', '2018-12-05', '2018-12-13',
                        '2018-12-19', '2018-12-24', '2018-12-27', '2019-01-08', '2019-01-11',
                        '2019-01-22', '2019-01-24', '2019-01-28', '2019-02-01', '2019-02-04',
                        '2019-02-07', '2019-02-12', '2019-02-15', '2019-02-21', '2019-03-03',
                        '2019-03-07', '2019-03-12', '2019-03-18', '2019-03-23' ]),
  'lower_window': 0,
  'upper_window': 5,
})


articles.head()
```

```python
prophet2 = Prophet(holidays=articles, daily_seasonality=True).fit(data)
future2 = prophet2.make_future_dataframe(periods=90)
forecast2 = prophet2.predict(future2)
```

```python
prophet2.plot(forecast2);
```

```python
prophet2.plot_components(forecast2, figsize=(10,15));
```

## Predicting Views for the next 15 days

```python
prophet3 = Prophet(holidays=articles, mcmc_samples=300).fit(data)
future3 = prophet3.make_future_dataframe(periods=60)
forecast3 = prophet3.predict(future3)
forecast3["Views"] = np.exp(forecast3.yhat).round()
forecast3["Views_lower"] = np.exp(forecast3.yhat_lower).round()
forecast3["Views_upper"] = np.exp(forecast3.yhat_upper).round()
forecast3[(forecast3.ds > "3-22-2019") &
          (forecast3.ds < "4-07-2019")][["ds","Views_lower",
                                        "Views", "Views_upper"]]
```

```python
prophet3.plot(forecast3);
```

## Exporting the results to CSV

```python
forecast3.to_csv("Predicted_Views", sep='\t', encoding='utf-8')
```
